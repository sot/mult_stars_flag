## Inject synthetic centroiding errors to study impact of multiple stars flag

This notebook provides limits on the impact of disabling the multiple stars
flag in the OBC star processing algorithm.  The idea is to inject synthetic
noise to simulate worst-case scenarios.  Comparing the before / after Kalman
filter output for the attitude and gyro-bias solution gives an measure of
the maximum errors due to the injected error.

The worse case scenario discussed here has bright background blobs on two corner
stars that perturb the star centroids by up to 2.5 arcsec in each axis.  This
then perturbs the aspect solution by a mean of around 140 arcsec in roll.  For a
pitch or yaw maneuver of around 90 degrees this would result in a maneuver
error of 140 arcsec and most likely a BSH / NSM safing event.  The next step
is to determine from telemetry if this case is credible and/or determine a
more realistic worst case.

The Kalman filter in this case is the ground aspect pipeline forward filter.
This is based on a TRW FORTRAN implementation of the flight filter from pre-launch.
At present there is no quantitivative comparison of the outputs from the
OBC to those from this ground implementation, though it is known that the
outputs are qualitatively similar.

```python
from astropy.table import Table
from astropy.io import fits
import matplotlib.pyplot as plt
from Ska.engarchive import fetch_eng as fetch
import glob
from mica.quaternion import Quat
%matplotlib inline
```

```python
def get_asp_files(obsid, case_id, prod, base_id):
    if isinstance(case_id, int):
        case_id = 'case{}'.format(case_id)
    filename = glob.glob('obs{}/{}/out1/pcadf*{}1.fits'.format(obsid, base_id, prod))[0]
    base = Table.read(filename)
    filename = glob.glob('obs{}/{}/out1/pcadf*{}1.fits'.format(obsid, case_id, prod))[0]
    case = Table.read(filename)
    maxtime = np.max(case['time'])
    # ok = base['time'] <= maxtime + 0.0001
    # base = base[ok]
    return base, case
```

```python
def compare_centroids_to_baseline(obsid, case_id, base_id):
    # Centroids
    base, case = get_asp_files(obsid, case_id, 'acen', base_id)
    t0 = base['time'][0]
    plt.figure(1, figsize=(12,4))
    for axis, iplot in (('y', 1), ('z', 2)):
        plt.subplot(1, 2, iplot)
        for slot in range(3, 8):
            cok = (case['slot'] == slot) & (case['alg'] == 8)
            bok = (base['slot'] == slot) & (base['alg'] == 8)
            if not np.any(cok):  # Slot might have been excluded
                continue
            ang = 'ang_{}_sm'.format(axis)
            plt.plot(base['time'][bok] - t0, (case[ang][cok] - base[ang][bok]) * 3600,
                     label='slot {}'.format(slot))
        plt.xlabel('time (sec)')
        plt.ylabel('d{} (arcsec)'.format(axis))
        plt.grid()
        plt.title('Star centroid differences')
    plt.legend(fontsize='small')
    plt.tight_layout()
```

```python
def compare_aspect_to_baseline(obsid, case_id, base_id):
    # Centroids
    rad2arcsec = np.degrees(1) * 3600
    base, case = get_asp_files(obsid, case_id, 'asol', base_id)
    base_q = Quat(base['q_att'])
    case_q = Quat(case['q_att'])
    dq = base_q.inv() * case_q  # case_q = base_q * dq
    # import pdb; pdb.set_trace()
    plt.figure(2, figsize=(12,4))
    plt.subplot(1, 2, 1)
    plt.plot(base['time'], 2 * rad2arcsec * dq.q[:, 0], label='roll')
    plt.plot(base['time'], 2 * rad2arcsec * dq.q[:, 1], label='pitch')
    plt.plot(base['time'], 2 * rad2arcsec * dq.q[:, 2], label='yaw')
    plt.grid()
    plt.title('Aspect solution difference')
    plt.legend(fontsize='small')

    plt.subplot(1, 2, 2)
    i0 = 20
    plt.plot(base['time'][i0:], (case['roll_bias'] - base['roll_bias'])[i0:] * 3600**2)
    plt.plot(base['time'][i0:], (case['pitch_bias'] - base['pitch_bias'])[i0:] * 3600**2)
    plt.plot(base['time'][i0:], (case['yaw_bias'] - base['yaw_bias'])[i0:] * 3600**2)
    plt.grid()
    plt.legend(fontsize='small')
    plt.title('Gyro bias difference')
    plt.tight_layout();
```

```python
def compare_forward_kalman_to_baseline(obsid, case_id, base_id):
    # Centroids
    rad2arcsec = np.degrees(1) * 3600
    base, case = get_asp_files(obsid, case_id, 'tklmn', base_id)
    times = base['time'] - base['time'][0]
    dx = case['xexdat'] - base['xexdat']
    plt.figure(3, figsize=(12,4))
    ax = plt.subplot(1, 2, 1)
    plt.plot(times, rad2arcsec * dx[:, 1], label='pitch')
    plt.plot(times, rad2arcsec * dx[:, 2], label='yaw')
    plt.plot(times, rad2arcsec * dx[:, 0], label='roll')
    plt.grid()
    plt.ylabel('Pitch and yaw angle (arcsec)')
    plt.title('Aspect solution difference')
    plt.xlabel('Time (sec)')
    plt.legend(fontsize='small')
 

    ax = plt.subplot(1, 2, 2)
    i0 = np.searchsorted(times, 1000)
    plt.plot(times[i0:], rad2arcsec * dx[i0:, 3], label='roll_bias')
    plt.plot(times[i0:], rad2arcsec * dx[i0:, 4], label='pitch_bias')
    plt.plot(times[i0:], rad2arcsec * dx[i0:, 5], label='yaw_bias')
    plt.xlim(0, None)
    ymax = np.max(np.abs(dx[i0:, 3:])) * rad2arcsec
    plt.ylim(-ymax, ymax)
    plt.grid()
    plt.ylabel('Bias rate (arcsec / sec)')
    plt.xlabel('Time (sec)')
    plt.title('Gyro bias difference')
    
    plt.tight_layout()
```

```python
def compare_all(obsid, case_id, base_id='baseline'):
    if isinstance(case_id, int):
        case_id = 'case{}'.format(case_id)
    with open('obs{}/{}/info'.format(obsid, case_id), 'r') as fh:
        print(fh.read())
    compare_centroids_to_baseline(obsid, case_id, base_id)
    compare_forward_kalman_to_baseline(obsid, case_id, base_id)
```

## [Obsid 2793](http://kadi.cfa.harvard.edu/mica/?obsid_or_date=2793)

Here we use obsid 2793, which is about 35 ksec and has a nice pattern of stars that allows inserting a worst-case roll error using star indices [6] and [8].  In addition data for star index [4] has been removed from processing to allow [6] to have an impact.

In the cases below only the first 5000 seconds of the observation are processed.

The only realistic case here is the **Error from two blobs** case below.

```python
compare_all(2793, 1, 'base5000')
```

#### Correlated noise for Y and Z in all slots with 10000 sec period

```python
compare_all(2793, 7)
```

#### Roll-inducing error in slots 5, 7 (index 6, 8)

This pattern of centroid offsets is not realistic for two blobs

### Error from two blobs

This is the most realistic case with a "spoiler" blob having 20% of the counts of the guide star and offset by dy,dz = 8, 8 arcsec.  Star indices [6] and [8] have such a spoiler blob in a way that maximixes the roll impact.

```python
compare_all(2793, 'multstar1', 'base5000')
```
